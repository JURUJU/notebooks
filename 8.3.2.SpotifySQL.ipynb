```python
albums = []
for artist in [bruce_sprinsteen, bob_marley]:
    results = sp.artist_albums(artist, album_type='album')
    albums.extend(results['items'])
    while results['next']:
        results = sp.next(results)
        albums.extend(results['items'])
```

```python
albums_df = pd.DataFrame([{'id':album['id'], 
 'name': album['name'],
 'cover': album['images'][0]['url'],
 'year':album['release_date'][:4],
 'artist':album['artists'][0]['id']} for album in albums])
```

```python
albums_df.to_sql('Album', db_url, index=False, if_exists='replace')
```

## Tracks

```python
track_list = []
for album in albums:
    results = sp.album_tracks(album['id'])
    for track in results['items']:  
        track_list.append( {
              'id': track['id'],
              'album':  album['id'],
              'duration': round(track['duration_ms']/1000),
              'name': track['name']
            }) 
    while results['next']:
        results = sp.next(results)
        for track in results['items']:  
            track_list.append( {
                  'id': track['id'],
                  'album':  album['id'],
                  'duration': round(track['duration_ms']/1000),
                  'name': track['name']
                }) 
```

```python
pd.DataFrame(track_list).to_sql('Track', db_url, index=False, if_exists='replace')
```
