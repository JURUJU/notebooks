```python
import requests
```

```python
base_prov_url = 'https://www.altimetrias.net/aspbk/ListadoPuertos.asp?Provincia='
```

```python
provincia = '01'
```

```python
url = f'{base_prov_url}{provincia}'
```

```python
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tabla = soup.find('table', attrs={'class':'txiki'})
```

```python
filas = tabla.findChildren('tr')
```
