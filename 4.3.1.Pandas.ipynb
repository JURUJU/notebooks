```python
nba = nba.dropna()
```

```python
def feet2cm(x):
    (f,i) = x.split('-')
    return 30.48 * int(f) + 2.54 * int(i)
```

```python
nba['altura_cm'] = nba.Height.apply(feet2cm)
```

## Merge

```python
nba
```

```python
altura_media = pd.DataFrame(nba.groupby('Team')['altura_cm'].mean())
```

```python
altura_media.reset_index(inplace=True)
```

```python
altura_media.columns = ['Equipo', 'Altura']
```

```python
altura_media = altura_media.rename(columns={'Altura':'altura_media_cm'})
```

```python
altura_media
```

```python
nba2 = pd.merge(nba, altura_media, on='Team')
```

```python
nba2
```

```python
nba2['estatura'] = np.where(nba2.altura_cm > 200, 'mas de dos metros', 'menos de dos metros')
```
